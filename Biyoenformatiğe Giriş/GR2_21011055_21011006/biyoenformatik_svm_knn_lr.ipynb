```python
import os
import numpy as np
import pandas as pd
```

```python
df_label = pd.read_csv(gdrivePath +os.sep+"Biyoenformatik"+os.sep+"GSE164191.csv",index_col=0)
df_label = df_label["status"]
df_label.replace({"normal": 0, "colorectal cancer": 1}, inplace=True)
df_label = pd.DataFrame(df_label)
df_label
```

```python
df_label.value_counts()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data=df_label, x='status')
plt.title('Distribution of Labels')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
df = df.drop(columns=['status'])
X_train, X_test, y_train, y_test = train_test_split(df, df_label, test_size=0.2, random_state=0, stratify = df_label)
#Bu hata df'den status sutununu cikardiktan sonra tekrar calistira bastigimdan kaynaklaniyor.
```

```python
print("X_train:", X_train.shape, "X_test", X_test.shape)
print("y_train:", y_train.shape, "y_test", y_test.shape)
```

```python
from sklearn import svm

clf = svm.SVC(kernel="linear")

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Macro Precision:",metrics.precision_score(y_test, y_pred,average='macro'))
print("Macro Recall:",metrics.recall_score(y_test, y_pred,average='macro'))
```

```python
target_names = ['normal', 'colorectal cancer']
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(clf, X_train, y_train, cv=10)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {'kernel':('linear', 'rbf'), 'C':[1, 10]}

svc = svm.SVC()


grid_obj = GridSearchCV(svc, parameters, cv=5)
grid_obj = grid_obj.fit(X_train, y_train)


best_parameters = grid_obj.best_params_
print(best_parameters)


best_clf = grid_obj.best_estimator_

y_pred = best_clf.predict(X_test)

print(classification_report(y_test, y_pred, target_names=['normal','colorectal cancer']))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred, labels=best_clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=best_clf.classes_)
disp.plot()
plt.show()
```
